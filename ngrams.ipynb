```python
from wordcloud import WordCloud

# , background_color="black"
wc = WordCloud(width=2400, height=1600, max_font_size=280).generate_from_frequencies(
    fs_freq
)
```

```python
import matplotlib.pyplot as plt

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```
